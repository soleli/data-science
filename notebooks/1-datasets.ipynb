## Descripción de variables de interés

**valence**: medida de 0,0 a 1,0 que refleja la positividad musical que transmite una pista. Las pistas con una valencia alta suenan más positivas (p. ej., felices, alegres, eufóricas), mientras que las pistas con una valencia baja suenan más negativas (p. ej., tristes, deprimidas, enojadas).

**year**: Año

**acousticness**: medida de 0.0 a 1.0 que detecta sonidos acústicos en una pista.

**artists**: nombre del artista

**danceability**: Medida de 0.0 a 1 que indica qué tan adecuada es una pista para bailar en función de una combinación de elementos musicales que incluyen tempo, estabilidad del ritmo y fuerza del compás. Un valor más cercano a 0,0 indica que una pista es menos bailable, y un valor más cercano a 1,0 indica que una pista es más bailable.

**duration_ms**: La duración de una pista medida en milisegundos.
 
 **energy**: : Una medida de 0,0 a 1,0 que refleja la intensidad de una pista. Las pistas enérgicas suelen ser rápidas, ruidosas y ruidosas.
 
 **id**: Número identificador de la pista
 
 **instrumentalness**: Medida de 0,0 a 1 que refleja hasta qué punto una pista no contiene vocalizaciones. Cuanto más cerca esté el valor de instrumentalidad de 1,0, mayor será la probabilidad de que la pista no contenga contenido vocal.
 
 **key**: Medida de 0 a 11 que indica la clave de una pista.

 **liveness**: medida de 0.0 a 1.0 que detecta la presencia de una audiencia en la grabación. Los valores de vivacidad más altos representan una mayor probabilidad de que la pista se interprete en vivo.
 
 **loudness**: medida de -60 y 0 que representa el volumen general de una pista. Este valor se mide en decibelios (dB). Los valores de sonoridad se promedian en toda la pista.
